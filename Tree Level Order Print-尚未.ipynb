# Instructor
# Deque
# Time & space complexity ==  O(N)

```python
import collections

class Node(object):
    
    def __init__(self, val=None):

        self.val = val 
        self.leftchild = None
        self.rightchild = None
         
def levelOrderPrint(tree):
    
    if not tree:
        
        return
    
    nodes = collections.deque([tree])
    
    currentCount, nextCount = 1, 0
    
    while len(nodes) != 0:
        
        currentNode= nodes.popleft()
        
        currentCount = currentCount - 1
        
        print (currentNode.val, end = " ")
        
        if currentNode.leftchild:
            
            nodes.append(currentNode.leftchild)
            
            nextCount = nextCount + 1
            
        if currentNode.rightchild:
            
            nodes.append(currentNode.rightchild)
            
            nextCount = nextCount + 1
            
        if currentCount == 0:
            
            #finished printing current level
            print ('\n')
            
            currentCount, nextCount = nextCount, 0
            
            
root = Node(1)
root.leftchild = Node(2)
root.rightchild = Node(3)
root.leftchild.leftchild = Node(4)
root.rightchild.leftchild = Node(5)
root.rightchild.rightchild = Node(6)

print (levelOrderPrint(root))  

%timeit levelOrderPrint(root)
```

# Recurseive
# Time complexity ==  O(N^2)

```python
class Node(object):
    
    def __init__(self, val=None):

        self.val = val 
        self.leftchild = None
        self.rightchild = None

def printGivenLevel(tree, level):
    # helper function that prints a given level of the tree
    if not tree:
        return 
    if level == 1:
        print (tree.val, end = "")
    else:
        printGivenLevel(tree.leftchild, level - 1)
        printGivenLevel(tree.rightchild, level - 1) 
 
def findTotalLevels(tree):
    # helper function that finds the total height of the tree
    if not tree:
        return 0
    else:
        
        left_levels = findTotalLevels(tree.leftchild)
        right_levels = findTotalLevels(tree.rightchild)
    return max(left_levels, right_levels) + 1
 
def levelOrderPrint(tree):
    # iterate through the entire tree and print each level
    levels = findTotalLevels(tree)
    
    for i in range(levels):
        printGivenLevel(tree, i+1) 
        print ('\n')
        
        
root = Node(1)
root.leftchild = Node(2)
root.rightchild = Node(3)
root.leftchild.leftchild = Node(4)
root.rightchild.leftchild = Node(5)
root.rightchild.rightchild = Node(6)

print (levelOrderPrint(root))  

%timeit levelOrderPrint(root)
```
